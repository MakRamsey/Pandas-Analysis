-Initial Set up:

```python
#Import dependencies needed for analysis
import pandas as pd
```

```python
#Establish variable to hold path for csv files
students_complete_csv = "Resources/students_complete.csv"
schools_complete_csv = "Resources/schools_complete.csv"
```

```python
#Create initial dataframe for both datasets
students_complete_df = pd.read_csv(students_complete_csv)
schools_complete_df = pd.read_csv(schools_complete_csv)
```

```python
#Visualize dataframes to confirm accuracy (Student dataset)
students_complete_df.head()
```

```python
#Visualize dataframes to confirm accuracy (School dataset)
schools_complete_df.head()
```

```python
#Combine both datasets into a single dataset and visualize the new dataframe
complete_data_df = pd.merge(students_complete_df, schools_complete_df, how="left", on=["school_name"])
complete_data_df.head()
```

-District Summary Analysis:

```python
#Calcualating the total number of schools (unique entries):
#Filtering the "School_Name" column for only unique values and storing that list within a unique variable for counting
unique_schools_list = complete_data_df["school_name"].unique()
#Counting number of entries, which represents total number of schools
school_count = len(unique_schools_list)
school_count
```

```python
#Calculating the total number of students:
#Filter the dataframe for the "Student ID" column and utilize the count function to return value for total number of students (initiate a variable to hold this value for future use as well)
student_count = complete_data_df["Student ID"].count()
student_count
```

```python
#Calculating the total budget:
#Initate a variable to hold a list of the unique entries from filtering our complete dataframe for unique values wtihn the "Budget" column
unique_budget_list = complete_data_df["budget"].unique()
#Utilize the sum function to generate the total budget for the district and store in a variable for future use
total_district_budget = unique_budget_list.sum()
total_district_budget
```

```python
#Calculating the average/mean math score:
#Utilize the describe function to provide a key stats breakdown inclusive of the global mean for both math and reading scores
complete_data_df.describe()
#Store this now verified dataframe
described_complete_df = complete_data_df.describe()
described_complete_df
```

```python
#Initiate variables to hold the average/mean values for math and reading scores derived from our new dataframe
average_math_score = described_complete_df.iloc[1,2]
average_math_score
#could have also used complete_data_df["math_score"].mean() to calculate mean math score
```

```python
average_reading_score = described_complete_df.iloc[1,1]
average_reading_score
#could have also used complete_data_df["reading_score"].mean() to calculate mean reading score
```

```python
#Calculating the percentage of students that passed math (defined as grade >= 70):
#First create a new filtered dataframe that only contains students with a math score of 70 and higher (passing) and count students within the list
students_math_passing_df = complete_data_df.loc[complete_data_df["math_score"] >= 70]
students_math_passing = students_math_passing_df["Student ID"].count()
#Find the percentage via dividing the number of students passing math (defined above) by the total number of students in the district (defined above) and multiplying by 100
percentage_math_passing = (students_math_passing / student_count) * 100
percentage_math_passing
#Please note this is the same as utilizing provided code 
```

```python
#Calculating the percentage of students that passed reading (defined as grade >= 70):
#First create a new filtered dataframe that only contains students with a reading score of 70 and higher (passing) and count students within
students_reading_passing_df = complete_data_df.loc[complete_data_df["reading_score"] >= 70]
students_reading_passing = students_reading_passing_df["Student ID"].count()
#Find the percentage via dividing the number of students passing reading (defined above) by the total number of students in the distric (defined above) and multiplying by 100
percentage_reading_passing = (students_reading_passing / student_count) * 100
percentage_reading_passing
#Please note this is the same as utilizing provided code 
```

```python
#Calculating the percentage of students who passed math and reading (defined as grades >= 70):
#First create a new filtered dataframe that only contains students with reading AND math scores of 70 and higher (passing) and count students within
students_both_passing_df = complete_data_df.loc[(complete_data_df["reading_score"] >= 70) & (complete_data_df["math_score"] >= 70)]
students_both_passing = students_both_passing_df["Student ID"].count()
#Finda the percentage via dividing the number of students passing reading & math (defined above) by the total number of students in the distric (defined above) and multiply by 100
percentage_both_passing = (students_both_passing / student_count) * 100
percentage_both_passing
#Please note this is the same as utilizing provided code 
```

```python
#Create a high-level snapshot of the districts key metrics in a dataframe
#Create dictionary containing the key metrics and load that data into a new dataframe
district_summary_dict = { "Total Number of Schools": [school_count], "Total Students": [student_count], "Total Budget": [total_district_budget],
                       "Average Math Score": [average_math_score], "Average Reading Score": [average_reading_score], "% Passing Math": [percentage_math_passing],
                       "% Passing Reading": [percentage_reading_passing], "% Overall Passing": [percentage_both_passing]}
district_summary_df = pd.DataFrame(district_summary_dict)
#Rename currrent row index value from '0' to 'Entire District:' for presentation value
district_summary_df = district_summary_df.rename(index={0: "Entire District:"})
#Format the resulting dataframe proper according to the data being represented
#I utilized numerical separators (,) for Total Students and Total Budget (along with a $ sign and two decimal places)
#Displayed Average Math/Reading score as well as the percentage values to two decimal places and included a % (for percentages only)
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:,.2f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:,.2f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:,.2f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:,.2f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:,.2f}".format)
district_summary_df
```

```python
#Include a new Per Student Budget column within our school summary dataframe which calculates the school's budget / number of students
school_summary_df["Per Student Budget"] = school_summary_df["budget"] / school_summary_df["size"]
school_summary_df
```

```python
#Create a groupedby dataset which groups our complete dataset by individual school name
#This will allow us to easily calculate desired metrics
grouped_school_stats = complete_data_df.groupby("school_name")
#Calculate the average/mean for each schools math scores and represent that data within a new dataframe for easy future merge
average_math_per_school = grouped_school_stats["math_score"].mean()
average_math_per_school_df = pd.DataFrame(average_math_per_school)
average_math_per_school_df
```

```python
#Calculate the average/mean for each schools reading scores and represent that data within a new dataframe for easy future merge
average_reading_per_school = grouped_school_stats["reading_score"].mean()
average_reading_per_school_df = pd.DataFrame(average_reading_per_school)
average_reading_per_school_df
```

```python
#Utilizing the above math/reading scores dataframes, merge them both into our master school summary dataframe via "school name"
merged_math_df = pd.merge(school_summary_df, average_math_per_school_df, on="school_name")
merged_both_df = pd.merge(merged_math_df, average_reading_per_school_df, on="school_name")
merged_both_df
```

```python
#Filter our comprehensive/complete dataframe to only inlcude students with passing math scores (greater than or equal to 70)
overall_math_pass = complete_data_df.loc[(complete_data_df["math_score"] >= 70)] 
#Group this filtered dataset by school name and utilize the size/count function to return a list of total students passing math for each school
grouped_overall_math_pass = overall_math_pass.groupby(["school_name"]).size()
#Create a dataframe utilizing a dictionary that has a new key "Students Passing Math" with corresponding values from the above list
grouped_overall_math_pass_df = pd.DataFrame({"Students Passing Math": grouped_overall_math_pass})
#Cast the students passing math values as floats for future calculations
grouped_overall_math_pass_df["Students Passing Math"].astype(float)
#Sort our master merged dataframe by school name in alphabetical order to align with new dataframes for joining and relative calculations
#This will be imperative for adding our reading passing percentage table as well
merged_both_df = merged_both_df.sort_values(["school_name"], ascending=True)
#Align our new dataframe's indexes with our sorted master merged dataframes indexes
grouped_overall_math_pass_df.index = merged_both_df.index
```

```python
#Include a new column "% Passing Math" to our master school summary dataframe via the following calculation
merged_both_df["% Passing Math"] = (grouped_overall_math_pass_df["Students Passing Math"] / merged_both_df["size"]) * 100
merged_both_df
```

```python
#Conduct a similiar order of operations in order to include a % Passing Reading column to our main school summary dataframe
#Filter our comprehensive/complete dataframe to only inlcude students with passing reading scores (greater than or equal to 70)
overall_reading_pass = complete_data_df.loc[(complete_data_df["reading_score"] >= 70)] 
#Group this filtered dataset by school name and utilize the size/count function to return a list of total students passing reading for each school
grouped_overall_reading_pass = overall_reading_pass.groupby(["school_name"]).size()
#Create a dataframe utilizing a dictionary that has a new key "Students Passing Reading" with corresponding values from the above list
grouped_overall_reading_pass_df = pd.DataFrame({"Students Passing Reading": grouped_overall_reading_pass})
#Cast the students passing reading values as floats for future calculations
grouped_overall_reading_pass_df["Students Passing Reading"].astype(float)
#Align this new dataframe's indexes with our master merged dataframes indexes
grouped_overall_reading_pass_df.index = merged_both_df.index
#Include a new column "% Passing Reading" to our master school summary dataframe via the following calculation
merged_both_df["% Passing Reading"] = (grouped_overall_reading_pass_df["Students Passing Reading"] / merged_both_df["size"]) * 100
merged_both_df
```

```python
#Conduct a similiar order of operations in order to include a % Overall Passing (math and reading) column to our main school summary dataframe
#Filter our comprehensive/complete dataframe to only inlcude students with passing reading AND math scores (greater than or equal to 70)
overall_both_pass = complete_data_df.loc[(complete_data_df["reading_score"] >= 70) & (complete_data_df["math_score"] >= 70)]
#Group this filtered dataset by school name and utilize the size/count function to return a list of total students passing reading for each school
grouped_overall_both_pass = overall_both_pass.groupby(["school_name"]).size()
#Create a dataframe utilizing a dictionary that has a new key "Students Passing Both" with corresponding values from the above list
grouped_overall_both_pass_df = pd.DataFrame({"Students Passing Both": grouped_overall_both_pass})
#Cast the students passing both values as floats for future calculations
grouped_overall_both_pass_df["Students Passing Both"].astype(float)
#Align this new dataframe's indexes with our master merged dataframes indexes
grouped_overall_both_pass_df.index = merged_both_df.index
#Include a new column "% Overall Passing" to our master school summary dataframe via the following calculation
merged_both_df["% Overall Passing"] = (grouped_overall_both_pass_df["Students Passing Both"] / merged_both_df["size"]) * 100
merged_both_df
##Please note this is the same as utilizing provided code snippet
```

```python
#Reformat the above school summary merged dataframe to diplay our key metrics appropriately
#Re-set the index values and rename the dataframe to Overall Per School Summary 
overall_per_school_summary = merged_both_df.reset_index(drop=True)
#Rename column headers appropriately
overall_per_school_summary = overall_per_school_summary.rename(columns = {"school_name": "School Name", "type": "School Type", "size": "Total Students", "budget": "Total School Budget", "math_score": "Average Math Score", "reading_score": "Average Reading Score"})
#Insert fomatting rules for each appropriate column
overall_per_school_summary["Total Students"] = overall_per_school_summary["Total Students"].map("{:,}".format)
overall_per_school_summary["Total School Budget"] = overall_per_school_summary["Total School Budget"].map("${:,.2f}".format)
overall_per_school_summary["Per Student Budget"] = overall_per_school_summary["Per Student Budget"].map("${:,.2f}".format)
overall_per_school_summary["Average Math Score"] = overall_per_school_summary["Average Math Score"].map("{:,.2f}".format)
overall_per_school_summary["Average Reading Score"] = overall_per_school_summary["Average Reading Score"].map("{:,.2f}".format)
overall_per_school_summary["% Passing Math"] = overall_per_school_summary["% Passing Math"].map("{:,.2f}".format)
overall_per_school_summary["% Passing Reading"] = overall_per_school_summary["% Passing Reading"].map("{:,.2f}".format)
overall_per_school_summary["% Overall Passing"] = overall_per_school_summary["% Overall Passing"].map("{:,.2f}".format)
overall_per_school_summary = overall_per_school_summary.set_index(["School Name"])
overall_per_school_summary
```

-Scores by School Spending:

```python
#Utilize a binning strategy in order to analyze scores by school spending
#Determine bins and labels (defined within challange prompt)
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#Create a copy of our 'overall per school summary' dataframe so we can manipulate/cut freely and because it has the per student budget category already populated
school_spending_df = overall_per_school_summary.copy()
```

```python
#Utilize pandas cut function to categorize spending (per student) based on the bins defined above and display new dataframe
#Please note, had to convert format of Per Student budget column back to integer for categorization
#First removed non integer symbol ($) and then converted to numeric (portion of this code assisted with Xpert Learning Assistant & Tutor)
#Remeber to reinsert "$" symbol if need to display later
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace("$", '')
school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"])

#Binning of data and display
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], spending_bins, labels=labels, include_lowest=True)
school_spending_df
```

```python
#Calculation of means for desired columns
#Reset the columns datatypes form objects to floats for mean calculation
school_spending_df = school_spending_df.astype({"Average Math Score": float,
                                                "Average Reading Score": float,
                                                "% Passing Math": float,
                                                "% Passing Reading": float,
                                                "% Overall Passing": float})
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=True)["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=True)["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=True)["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=True)["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"], observed=True)["% Overall Passing"].mean()
```

```python
#Assemblying the above into a new dataframe
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing": overall_passing_spending})
#Display resulting dataframe
spending_summary
#Please note very slight discrepancy between my results (decimals) and the starter code's results due to previously rounded/formatted data within the per school summary dataframe
```
